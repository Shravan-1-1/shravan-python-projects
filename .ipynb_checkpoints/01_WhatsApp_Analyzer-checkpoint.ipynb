```python
print("üì± DAY 5 - PROBLEM 1: WHATSAPP CHAT ANALYZER")
print("=" * 60)

# Fake WhatsApp chat data (Real file se bhi kar sakte ho)
whatsapp_chat = """Shravan: Good morning everyone! 
Amit: GM bhai, kaisa hai?
Shravan: Mast hu, Python practice kar raha
Rahul: Day 5 start kiya?
Shravan: Haan bhai, WhatsApp analyzer ban raha!
Priya: üî• Boss level project!
Shravan: Thanks! Ab stats nikalenge
Amit: Word count kitna?
Shravan: Bas 5 minute mein ready!"""

# 1. Messages split karo
messages = whatsapp_chat.split('\n')
shravan_count = 0
total_messages = len(messages)

for msg in messages:
    if msg.startswith('Shravan:'):
        shravan_count += 1

# 2. Words count
shravan_words = whatsapp_chat.lower().count('shravan')
total_words = len(whatsapp_chat.split())

print(f"\nüìä SHRAVAN WHATSAPP STATS")
print("=" * 40)
print(f"Total Messages: {total_messages}")
print(f"Shravan Messages: {shravan_count}")
print(f"Shravan %: {(shravan_count/total_messages)*100:.1f}%")
print(f"Total Words: {total_words}")
print(f"'Shravan' word: {shravan_words} times")

# 3. Top words (Simple)
words = whatsapp_chat.lower().split()
top_words = {}
for word in words:
    word = word.strip('!?,.')
    if len(word) > 3:
        top_words[word] = top_words.get(word, 0) + 1

print("\nüèÜ TOP 5 WORDS:")
sorted_words = sorted(top_words.items(), key=lambda x: x[1], reverse=True)[:5]
for i, (word, count) in enumerate(sorted_words, 1):
    print(f"{i}. '{word}' ‚Üí {count} times")
```
