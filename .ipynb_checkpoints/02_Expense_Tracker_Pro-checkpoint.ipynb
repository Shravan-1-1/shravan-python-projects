```python
print("ðŸ’° DAY 5 - PROBLEM 2: EXPENSE TRACKER PRO")
print("=" * 60)

import json
from datetime import datetime

class ExpenseTracker:
    def __init__(self):
        self.expenses = []
        self.load_expenses()
    
    def load_expenses(self):
        """JSON se purane expenses load karo"""
        try:
            with open("shravan_expenses.json", "r", encoding="utf-8") as f:
                self.expenses = json.load(f)
            print(f"âœ… {len(self.expenses)} expenses loaded!")
        except:
            print("ðŸ†• New tracker started!")
    
    def save_expenses(self):
        """Permanent save"""
        with open("shravan_expenses.json", "w", encoding="utf-8") as f:
            json.dump(self.expenses, f, indent=2, ensure_ascii=False)
    
    def add_expense(self, category, amount, desc=""):
        """Naya expense add karo"""
        expense = {
            "date": datetime.now().strftime("%Y-%m-%d"),
            "category": category,
            "amount": amount,
            "desc": desc
        }
        self.expenses.append(expense)
        self.save_expenses()
        print(f"âœ… â‚¹{amount} ({category}) added!")
    
    def monthly_report(self):
        """Pro level monthly stats"""
        if not self.expenses:
            print("ðŸ“­ No expenses yet!")
            return
        
        today = datetime.now()
        month_start = today.replace(day=1)
        month_expenses = [e for e in self.expenses if datetime.strptime(e['date'], "%Y-%m-%d") >= month_start]
        
        total_spent = sum(e['amount'] for e in month_expenses)
        categories = {}
        
        for exp in month_expenses:
            cat = exp['category']
            categories[cat] = categories.get(cat, 0) + exp['amount']
        
        print(f"\nðŸ“Š SHRAVAN MONTHLY REPORT ({today.strftime('%B %Y')})")
        print("=" * 50)
        print(f"ðŸ’¸ Total Spent: â‚¹{total_spent:,.0f}")
        print(f"ðŸ“… Days: {len(set(e['date'] for e in month_expenses))}")
        print(f"ðŸ“¦ Transactions: {len(month_expenses)}")
        
        print("\nðŸ† CATEGORY BREAKDOWN:")
        for cat, amt in sorted(categories.items(), key=lambda x: x[1], reverse=True):
            percent = (amt/total_spent)*100
            print(f"  {cat:>10}: â‚¹{amt:>8,.0f} ({percent:>5.1f}%)")

# ðŸ”¥ LIVE DEMO - Shravan ka real expenses
print("\n" + "â”€"*70)
tracker = ExpenseTracker()

# Demo expenses (Real life)
tracker.add_expense("Food", 250, "Lunch + Chai")
tracker.add_expense("Transport", 100, "Cycle repair")
tracker.add_expense("Study", 300, "Python course")
tracker.add_expense("Food", 180, "Dinner")
tracker.add_expense("Entertainment", 150, "Netflix")

# Pro report generate
tracker.monthly_report()
```
