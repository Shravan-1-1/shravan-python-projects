```python
print("ğŸš— DAY 3 - PROBLEM 3: CAR SHOWROOM MANAGEMENT")
print("=" * 60)

class Car:
    def __init__(self, model, company, year, price, mileage, color):
        self.model = model
        self.company = company
        self.year = year
        self.price = price
        self.mileage = mileage
        self.color = color
        self.status = "Available"  # Available/Sold
    
    def display_info(self):
        return f"""
ğŸš— {self.company} {self.model}
Year: {self.year} | Color: {self.color}
ğŸ’° Price: â‚¹{self.price:,.0f} | â›½ Mileage: {self.mileage} kmpl
Status: {self.status}
        """.strip()
    
    def sell_car(self):
        if self.status == "Available":
            self.status = "Sold"
            return f"âœ… {self.company} {self.model} SOLD!"
        return "âŒ Already sold!"

# ğŸ”¥ 10 CAR SHOWROOM INVENTORY
showroom = [
    Car("Swift", "Maruti", 2024, 750000, 22.5, "Red"),
    Car("Creta", "Hyundai", 2023, 1450000, 18.0, "White"),
    Car("XUV700", "Mahindra", 2024, 1850000, 16.5, "Black"),
    Car("Thar", "Mahindra", 2023, 1650000, 15.2, "Grey"),
    Car("i20", "Hyundai", 2024, 850000, 20.0, "Blue"),
    Car("Baleno", "Maruti", 2024, 820000, 23.0, "Silver"),
    Car("Seltos", "Hyundai", 2023, 1350000, 17.5, "Red"),
    Car("Scorpio", "Mahindra", 2022, 1250000, 14.5, "White"),
    Car("Brezza", "Maruti", 2024, 950000, 20.5, "Black"),
    Car("Nexon", "Tata", 2024, 1050000, 19.0, "Blue")
]

print(f"ğŸª SHRAVAN'S CAR SHOWROOM: {len(showroom)} cars ready!")

# ğŸ”¥ SMART SEARCH FUNCTIONS
def search_by_budget(max_price):
    return [car for car in showroom if car.price <= max_price and car.status == "Available"]

def search_by_company(company):
    return [car for car in showroom if company.lower() in car.company.lower() and car.status == "Available"]

def search_by_mileage(min_mileage):
    return [car for car in showroom if car.mileage >= min_mileage and car.status == "Available"]

# ğŸ”¥ DEMO - Shravan ka showroom test
print("\n" + "â”€"*70)
print("ğŸ”¥ SHRAVAN KUMAR SHARMA - SHOWROOM DEMO")
print("â”€"*70)

# 1. Budget search (under 10 lakh)
print("\nğŸ’° Cars under â‚¹10 Lakh:")
budget_cars = search_by_budget(1000000)
for i, car in enumerate(budget_cars[:3], 1):
    print(f"{i}. {car.model}")
    print(car.display_info())
    print()

# 2. Company search (Hyundai)
print("ğŸ† Hyundai cars:")
hyundai_cars = search_by_company("Hyundai")
for car in hyundai_cars:
    print(car.display_info())
    print()

# 3. Sell Swift (Test)
swift = next(car for car in showroom if car.model == "Swift")
print(swift.sell_car())
print(f"Swift status: {swift.status}")

# 4. Showroom summary
available = len([c for c in showroom if c.status == "Available"])
sold = len(showroom) - available
print(f"\nğŸ“Š SUMMARY: {available} Available | {sold} Sold")
```
