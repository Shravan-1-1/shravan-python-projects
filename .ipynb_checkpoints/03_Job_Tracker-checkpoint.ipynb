```python
print("ğŸ’¼ DAY 5 - PROBLEM 3: JOB APPLICATION TRACKER")
print("=" * 60)

import json
from datetime import datetime, timedelta

class JobTracker:
    def __init__(self):
        self.jobs = []
        self.load_jobs()
    
    def load_jobs(self):
        """Purane applications load karo"""
        try:
            with open("shravan_jobs.json", "r", encoding="utf-8") as f:
                self.jobs = json.load(f)
            print(f"âœ… {len(self.jobs)} job applications loaded!")
        except:
            print("ğŸ†• New job tracker started!")
    
    def save_jobs(self):
        """Permanent save"""
        with open("shravan_jobs.json", "w", encoding="utf-8") as f:
            json.dump(self.jobs, f, indent=2, ensure_ascii=False)
    
    def add_application(self, company, role, status="Applied", deadline="2026-03-15"):
        """Naya job application add karo"""
        job = {
            "company": company,
            "role": role,
            "status": status,
            "apply_date": datetime.now().strftime("%Y-%m-%d"),
            "deadline": deadline,
            "notes": ""
        }
        self.jobs.append(job)
        self.save_jobs()
        print(f"âœ… {company} - {role} application added!")
    
    def urgent_alerts(self):
        """Deadline ke paas wale jobs highlight karo"""
        alerts = []
        today = datetime.now().date()
        
        for job in self.jobs:
            deadline = datetime.strptime(job['deadline'], "%Y-%m-%d").date()
            days_left = (deadline - today).days
            
            if days_left <= 3 and job['status'] != "Interview":
                alerts.append((job, days_left))
        
        return alerts
    
    def status_report(self):
        """Complete job status dashboard"""
        if not self.jobs:
            print("ğŸ“­ No applications yet!")
            return
        
        statuses = {}
        for job in self.jobs:
            status = job['status']
            statuses[status] = statuses.get(status, 0) + 1
        
        print(f"\nğŸ“Š SHRAVAN JOB STATUS REPORT")
        print("=" * 50)
        print(f"ğŸ’¼ Total Applications: {len(self.jobs)}")
        
        for status, count in statuses.items():
            bar = "â–ˆ" * (count * 2)
            print(f"  {status:<12}: {count:>2} {bar}")
        
        # Urgent alerts
        alerts = self.urgent_alerts()
        if alerts:
            print("\nğŸš¨ URGENT ALERTS:")
            for job, days in alerts:
                print(f"  âš ï¸  {job['company']} ({days} days left)")

# ğŸ”¥ LIVE DEMO - Shravan ke real job applications
print("\n" + "â”€"*70)
tracker = JobTracker()

# Govt Job + AI jobs (Real applications)
tracker.add_application("SBI", "Assistant Manager", "Applied", "2026-03-05")
tracker.add_application("RBI", "Grade B Officer", "Test Cleared", "2026-03-10")
tracker.add_application("Google", "Python Developer", "Interview Round 1")
tracker.add_application("Microsoft", "AI Engineer", "Applied")
tracker.add_application("RRB", "Section Controller", "Exam Date Allotted", "2026-03-02")

# Complete dashboard
tracker.status_report()
```
