```python
print("ðŸ§  DAY 5 - PROBLEM 4: SHRAVAN PYTHON QUIZ")
print("=" * 60)

import json
import time
from datetime import datetime

class QuizGame:
    def __init__(self):
        self.score = 0
        self.total_questions = 0
        self.player_stats = []
        self.questions = [
            {
                "q": "Python mein list ka last element kaise nikale?",
                "options": ["a) list[-1]", "b) list.last()", "c) list.pop()", "d) list.end()"],
                "ans": "a"
            },
            {
                "q": "OOP mein 'self' ka matlab?",
                "options": ["a) Current object", "b) Class name", "c) Global variable", "d) Function"],
                "ans": "a"
            },
            {
                "q": "JSON file save karne ka function?",
                "options": ["a) json.dump()", "b) json.save()", "c) json.write()", "d) json.export()"],
                "ans": "a"
            },
            {
                "q": "'with open()' ka fayda?",
                "options": ["a) Auto file close", "b) Fast reading", "c) Memory save", "d) All"],
                "ans": "a"
            },
            {
                "q": "Error handling ka keyword?",
                "options": ["a) try/except", "b) error/catch", "c) if/error", "d) check/fix"],
                "ans": "a"
            }
        ]
    
    def start_quiz(self):
        """Quiz start with 30 sec timer per question"""
        print("\nðŸŽ® SHRAVAN PYTHON QUIZ START!")
        print("â±ï¸  Har question: 30 seconds | 5 Questions")
        print("Correct answer type karo (a/b/c/d)")
        
        self.score = 0
        self.total_questions = len(self.questions)
        
        for i, q in enumerate(self.questions, 1):
            print(f"\nâ”€{'â”€'*40}")
            print(f"Q{i}) {q['q']}")
            for opt in q['options']:
                print(f"   {opt}")
            
            # 30 sec timer
            start_time = time.time()
            user_ans = input(f"\nâ° Answer (a/b/c/d): ").strip().lower()
            time_taken = time.time() - start_time
            
            if time_taken > 30:
                print("â° Time out! Sahi jawab tha:", q['ans'].upper())
            elif user_ans == q['ans']:
                self.score += 1
                print("âœ… Correct! ðŸ”¥")
            else:
                print(f"âŒ Wrong! Sahi jawab: {q['ans'].upper()}")
            
            print(f"â±ï¸  Time: {time_taken:.1f} sec")
        
        self.show_result()
    
    def show_result(self):
        """Pro level result + JSON save"""
        percentage = (self.score / self.total_questions) * 100
        
        result = {
            "date": datetime.now().strftime("%Y-%m-%d %H:%M"),
            "score": self.score,
            "total": self.total_questions,
            "percentage": round(percentage, 1),
            "grade": "A+" if percentage >= 90 else "A" if percentage >= 80 else "B" if percentage >= 70 else "C"
        }
        
        self.player_stats.append(result)
        self.save_stats()
        
        print(f"\nðŸ† FINAL RESULT")
        print("=" * 40)
        print(f"Score: {self.score}/{self.total_questions}")
        print(f"Percentage: {percentage:.1f}%")
        print(f"Grade: {result['grade']}")
        print(f"Status: {'PYTHON MASTER! ðŸ”¥' if percentage >= 80 else 'Practice more! ðŸ’ª'}")
        
        # Best score
        if self.player_stats:
            best = max(self.player_stats, key=lambda x: x['percentage'])
            print(f"\nðŸ“Š Best Score: {best['score']}/{best['total']} ({best['percentage']}%)")
    
    def save_stats(self):
        """Permanent stats save"""
        with open("shravan_quiz_stats.json", "w", encoding="utf-8") as f:
            json.dump(self.player_stats, f, indent=2, ensure_ascii=False)

# ðŸ”¥ QUIZ START
print("\n" + "â”€"*70)
game = QuizGame()
game.start_quiz()
```
