```python
print("ğŸ¦¸â€â™‚ï¸ DAY 3 - PROBLEM 5: SHRAVAN SUPERHERO")
print("=" * 60)

# Base Class (Parent)
class Character:
    def __init__(self, name, power_level=100, health=100):
        self.name = name
        self.power_level = power_level
        self.health = health
        self.level = 1
        print(f"âš”ï¸ {name} character created! Power: {power_level}")
    
    def attack(self, enemy):
        damage = self.power_level // 10
        print(f"âš¡ {self.name} attacks {enemy}! Damage: {damage}")
        return damage
    
    def level_up(self):
        self.level += 1
        self.power_level += 20
        print(f"ğŸ® {self.name} leveled up! Level {self.level} | Power {self.power_level}")

# Superhero (Child Class) - Inheritance ğŸ”¥
class Superhero(Character):
    def __init__(self, name, superpower, power_level=150):
        super().__init__(name, power_level)  # Parent ko call
        self.superpower = superpower
        self.weapons = []
    
    def super_attack(self, enemy):
        damage = self.power_level // 5  # Double damage!
        print(f"ğŸŒŸ SHR AVAN uses {self.superpower}! {enemy} -{damage}HP!")
        return damage * 2
    
    def add_weapon(self, weapon):
        self.weapons.append(weapon)
        print(f"ğŸ”« {self.name} equipped {weapon}")

# ğŸ”¥ SHR AVAN SUPERHERO DEMO
print("\n" + "â”€"*70)

# Normal character
villain = Character("Dark Lord", 80, 120)

# SHR AVAN Superhero (Inheritance power!)
shravan = Superhero("Shravan", "Lightning Bolt", 200)
shravan.add_weapon("Laser Sword")
shravan.add_weapon("Plasma Gun")

print("\n" + "â”€"*70)
print("ğŸ¥· BATTLE START!")

# Normal attack
normal_damage = shravan.attack("Dark Lord")
villain.health -= normal_damage

# SUPER ATTACK ğŸ”¥
super_damage = shravan.super_attack("Dark Lord")
villain.health -= super_damage

# Level up
shravan.level_up()

# Final status
print(f"\nğŸ† FINAL STATUS:")
print(f"Shravan: Level {shravan.level} | Power {shravan.level} | Weapons: {len(shravan.weapons)}")
print(f"Dark Lord Health: {villain.health}/120 {'ğŸ’€ DEFEATED!' if villain.health <= 0 else ''}")
print(f"Superpower: {shravan.superpower}")
```
