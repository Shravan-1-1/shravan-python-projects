```python
print("ğŸ’¼ DAY 4 - PROBLEM 5: SHRAVAN PORTFOLIO APP")
print("=" * 60)

import json
from datetime import datetime

class ShravanPortfolio:
    def __init__(self):
        self.name = "Shravan"
        self.role = "Python Developer"
        self.projects = []
        self.skills = ["Python", "OOP", "File Handling", "JSON", "CSV", "Error Handling"]
        self.load_data()
    
    def load_data(self):
        """JSON se purana data load karo"""
        try:
            with open("shravan_portfolio.json", "r", encoding="utf-8") as f:
                data = json.load(f)
                self.projects = data.get("projects", [])
            print("âœ… Portfolio data loaded!")
        except FileNotFoundError:
            print("ğŸ†• New portfolio created!")
    
    def save_data(self):
        """JSON mein permanent save"""
        portfolio_data = {
            "name": self.name,
            "role": self.role,
            "skills": self.skills,
            "projects": self.projects,
            "last_updated": datetime.now().strftime("%Y-%m-%d %H:%M")
        }
        with open("shravan_portfolio.json", "w", encoding="utf-8") as f:
            json.dump(portfolio_data, f, indent=2, ensure_ascii=False)
        print("ğŸ’¾ Portfolio saved!")
    
    def add_project(self, name, tech, status="Completed"):
        """Naya project add karo"""
        project = {
            "name": name,
            "tech_stack": tech,
            "status": status,
            "date": datetime.now().strftime("%Y-%m-%d")
        }
        self.projects.append(project)
        print(f"âœ… '{name}' project added!")
    
    def show_portfolio(self):
        """Professional portfolio display"""
        total_projects = len(self.projects)
        completed = len([p for p in self.projects if p["status"] == "Completed"])
        
        print(f"\n{'='*60}")
        print(f"ğŸ’¼ SHRAVAN KA PORTFOLIO")
        print(f"{'='*60}")
        print(f"ğŸ‘¤ Name: {self.name:^20} | Role: {self.role}")
        print(f"ğŸ› ï¸  Skills: {', '.join(self.skills)}")
        print(f"ğŸ“Š Projects: {total_projects} | Completed: {completed}")
        print(f"ğŸ“… Last Updated: {datetime.now().strftime('%Y-%m-%d %H:%M')}")
        
        if self.projects:
            print("\nğŸ“‹ PROJECTS:")
            print("-" * 40)
            for i, proj in enumerate(self.projects[-5:], 1):  # Last 5
                status_emoji = "âœ…" if proj["status"] == "Completed" else "â³"
                print(f"{i}. {status_emoji} {proj['name']}")
                print(f"   Tech: {', '.join(proj['tech_stack'])}")
                print()
    
    def resume_ready(self):
        """GitHub/Resume ke liye clean text"""
        return f"""
SHRAVAN - PYTHON DEVELOPER
Prayagraj, UP | Govt Job + AI Developer Aspirant

SKILLS:
{', '.join(self.skills)}

PROJECTS ({len(self.projects)}):
{chr(10).join([f"- {p['name']} ({', '.join(p['tech_stack'])})" for p in self.projects[-3:]])}

GITHUB READY PORTFOLIO!
        """.strip()

# ğŸ”¥ SHRAVAN PORTFOLIO LIVE DEMO
print("\n" + "â”€"*70)

# Portfolio app start
portfolio = ShravanPortfolio()

# Real projects add karo (Day 1-4)
portfolio.add_project("Smart Bank", ["OOP", "JSON", "File Handling"])
portfolio.add_project("Car Showroom", ["OOP", "List Comprehension"])
portfolio.add_project("Student Database", ["CSV", "File Handling"])
portfolio.add_project("Error-Proof Calculator", ["Try/Except", "Functions"])

# Portfolio show
portfolio.show_portfolio()

# Save permanently
portfolio.save_data()

# Resume ready text
print("\n" + "â”€"*70)
print("ğŸ“„ RESUME READY TEXT (Copy karo):")
print(portfolio.resume_ready())
```
