```python
print("ğŸ¦ BANK ACCOUNT SYSTEM")
print("=" * 50)

# Account setup
balance = 1000.00
transactions = []

print(f"\nğŸ’³ Welcome! Current Balance: â‚¹{balance:.2f}")

while True:
    print("\nğŸ“‹ Choose option:")
    print("1. Check Balance")
    print("2. Deposit")
    print("3. Withdraw") 
    print("4. Transaction History")
    print("5. Exit")
    
    choice = input("Enter choice (1-5): ")
    
    if choice == '1':
        print(f"\nğŸ’° Current Balance: â‚¹{balance:.2f}")
        
    elif choice == '2':
        amount = float(input("Enter deposit amount: â‚¹"))
        if amount > 0:
            balance += amount
            transactions.append(f"Deposited: +â‚¹{amount}")
            print(f"âœ… Deposited â‚¹{amount:.2f}")
            print(f"ğŸ’³ New Balance: â‚¹{balance:.2f}")
        else:
            print("âŒ Invalid amount!")
            
    elif choice == '3':
        amount = float(input("Enter withdraw amount: â‚¹"))
        if amount > 0 and amount <= balance:
            balance -= amount
            transactions.append(f"Withdrew: -â‚¹{amount}")
            print(f"âœ… Withdrew â‚¹{amount:.2f}")
            print(f"ğŸ’³ New Balance: â‚¹{balance:.2f}")
        else:
            print("âŒ Insufficient balance or invalid amount!")
            
    elif choice == '4':
        print("\nğŸ“œ TRANSACTION HISTORY:")
        print("-" * 30)
        if transactions:
            for i, trans in enumerate(transactions[-10:], 1):  # Last 10
                print(f"{i}. {trans}")
        else:
            print("No transactions yet!")
        print(f"ğŸ’° Current Balance: â‚¹{balance:.2f}")
        
    elif choice == '5':
        print(f"\nğŸ‘‹ Thank you! Final Balance: â‚¹{balance:.2f}")
        break
        
    else:
        print("âŒ Invalid choice! Try 1-5")
```
