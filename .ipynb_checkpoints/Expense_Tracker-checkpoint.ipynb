```python
print("üí∞ DAY 2 - PROBLEM 3: EXPENSE TRACKER")
print("=" * 50)

import json
import os
EXPENSE_FILE = "expenses.json"

# Load expenses from file
def load_expenses():
    if os.path.exists(EXPENSE_FILE):
        try:
            with open(EXPENSE_FILE, 'r') as f:
                return json.load(f)
        except:
            return []
    return []

# Save expenses to file
def save_expenses(expenses):
    with open(EXPENSE_FILE, 'w') as f:
        json.dump(expenses, f, indent=2)

# Main program
expenses = load_expenses()
print(f"üìä Loaded {len(expenses)} previous expenses!")

while True:
    print("\n" + "="*30)
    print("1. Add Expense  2. Show Total  3. Show All  4. Exit")
    choice = input("Choose (1-4): ").strip()
    
    if choice == '1':
        category = input("Category (Food/Travel/Study/Other): ").title()
        amount = float(input("Amount (‚Çπ): "))
        desc = input("Description: ")
        
        expense = {
            "category": category,
            "amount": amount,
            "desc": desc,
            "date": "2026-02-26"  # Today's date
        }
        expenses.append(expense)
        save_expenses(expenses)
        print(f"‚úÖ Added ‚Çπ{amount} ({category}) - Saved!")
        
    elif choice == '2':
        total = sum(exp['amount'] for exp in expenses)
        print(f"üí∏ TOTAL EXPENSES: ‚Çπ{total:.2f}")
        
    elif choice == '3':
        if expenses:
            print("\nüìã ALL EXPENSES:")
            for i, exp in enumerate(expenses, 1):
                print(f"{i}. {exp['category']}: ‚Çπ{exp['amount']} - {exp['desc']}")
        else:
            print("üì≠ No expenses yet!")
            
    elif choice == '4':
        print("üëã Thanks! Data saved in expenses.json")
        break
        
    else:
        print("‚ùå Invalid! Choose 1-4")
        # COMPLETE! Expense tracker + JSON file save/load boss level‚Äîperfect! 
```
