```python
print("ðŸ’¾ DAY 4 - PROBLEM 2: JSON DATA MANAGER")
print("=" * 50)

import json

# 1. JSON FILE SAVE (Shravan ka data permanent)
shravan_data = {
    "name": "Shravan",
    "city": "Prayagraj",
    "skills": ["Python", "OOP", "File Handling"],
    "projects": {
        "Day3": ["Smart Bank", "Car Showroom", "Library"],
        "Day4": ["File Basics", "JSON Manager"]
    },
    "target": "Govt Job + AI Developer"
}

# SAVE JSON file
with open("shravan_profile.json", "w", encoding="utf-8") as f:
    json.dump(shravan_data, f, indent=2, ensure_ascii=False)
print("âœ… shravan_profile.json SAVE ho gaya!")

# 2. JSON FILE LOAD (Permanent data wapas laana)
print("\nðŸ“‚ Shravan ka JSON data:")
with open("shravan_profile.json", "r", encoding="utf-8") as f:
    loaded_data = json.load(f)
    print(f"Name: {loaded_data['name']}")
    print(f"Skills: {', '.join(loaded_data['skills'])}")
    print(f"Projects: {len(loaded_data['projects']['Day3'])} completed!")

# 3. UPDATE JSON (Naya data add karna)
loaded_data["skills"].append("JSON Handling")
loaded_data["projects"]["Day4"].append("JSON Manager")

# Update & Save
with open("shravan_profile.json", "w", encoding="utf-8") as f:
    json.dump(loaded_data, f, indent=2, ensure_ascii=False)
print("\nâž• New skills + projects UPDATE ho gaye!")

# 4. Pretty Print JSON
print("\nðŸ“‹ COMPLETE Shravan Profile (Updated):")
print(json.dumps(loaded_data, indent=2, ensure_ascii=False))
```

```python
# json.dump()  = Python â†’ JSON file
# json.load()  = JSON file â†’ Python  
# indent=2     = Pretty formatting (padhaai ke liye)
# ensure_ascii=False = Hindi support
```
