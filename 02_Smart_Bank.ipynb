```python
print("ðŸ¦ DAY 3 - PROBLEM 2: SMART BANKING SYSTEM")
print("=" * 60)

class SmartBank:
    def __init__(self, name, account_no, initial_balance=0):
        self.name = name
        self.account_no = account_no
        self.balance = initial_balance
        self.transaction_history = []
        self.loan_status = "Clear"
        self.vip_customer = initial_balance >= 50000
        print(f"ðŸ’Ž SMART ACCOUNT '{name}' ACTIVE | VIP: {self.vip_customer}")
    
    def deposit(self, amount, source="Cash"):
        if amount > 0:
            self.balance += amount
            self.transaction_history.append({
                "type": "CREDIT", 
                "amount": amount,
                "source": source,
                "time": "2026-02-26 15:39"
            })
            
            bonus = amount * 0.005 if self.vip_customer else 0
            if bonus > 0:
                self.balance += bonus
                print(f"ðŸŽ VIP BONUS: â‚¹{bonus:.2f}")
            
            print(f"âœ… â‚¹{amount:,} deposit | BALANCE: â‚¹{self.balance:,.0f}")
        else:
            print("âŒ Positive amount daalo!")
    
    def withdraw(self, amount, purpose="Personal"):
        if amount > 0 and amount <= self.balance:
            self.balance -= amount
            self.transaction_history.append({
                "type": "DEBIT", 
                "amount": amount,
                "purpose": purpose,
                "time": "2026-02-26 15:39"
            })
            
            if self.balance < 1000:
                print("âš ï¸ Low balance! Reload karo.")
            
            print(f"âœ… â‚¹{amount:,} withdraw | BALANCE: â‚¹{self.balance:,.0f}")
        else:
            print("âŒ Paise kam ya invalid amount!")
    
    def loan_request(self, amount):
        if self.loan_status == "Clear" and self.balance >= 5000:
            self.loan_status = "Pending"
            print(f"ðŸ’° Loan â‚¹{amount:,} request ACCEPTED!")
        else:
            print("âŒ Loan reject!")
    
    def dashboard(self):
        total_credit = sum(t["amount"] for t in self.transaction_history if t["type"]=="CREDIT")
        total_debit = sum(t["amount"] for t in self.transaction_history if t["type"]=="DEBIT")
        
        return f"""
{'ðŸ’Ž VIP DASHBOARD' if self.vip_customer else 'ðŸ¦ BANK DASHBOARD'}
â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•
ðŸ‘¤ {self.name:^20} | A/C: {self.account_no}
ðŸ’° Balance: â‚¹{self.balance:,.0f} {'â­VIP' if self.vip_customer else ''}
ðŸ“Š Credit: â‚¹{total_credit:,.0f} | Debit: â‚¹{total_debit:,.0f}

ðŸ“œ LAST 3 TRANSACTIONS:
{chr(10).join([f'  {t["type"][0]} â‚¹{t["amount"]:>8,} {t.get("source","")+t.get("purpose","")}' 
               for t in self.transaction_history[-3:]]) or '  No transactions'}

ðŸ’³ Loan: {self.loan_status}
        """.strip()

# ðŸ”¥ SHR AVAN ka VIP account
print("\n" + "â”€"*70)

shravan_bank = SmartBank("Shravan Singh", "SBIâ˜…VIP123", 55000)
shravan_bank.deposit(15000, "Salary")
shravan_bank.withdraw(8000, "Food+Cycle")
shravan_bank.loan_request(50000)
print(shravan_bank.dashboard())

print("\n" + "â”€"*70)

amit_bank = SmartBank("Amit Sharma", "HDFC001", 2500)
amit_bank.deposit(7000, "Freelance")
print(amit_bank.dashboard())
```
