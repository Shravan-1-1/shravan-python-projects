```python
print("ğŸ›¡ï¸ DAY 4 - PROBLEM 3: SMART CALCULATOR")
print("=" * 50)

def safe_input(prompt):
    """User se safe input leta hai - koi error nahi!"""
    while True:
        try:
            return float(input(prompt))
        except ValueError:
            print("âŒ Number daalo bhai! Text mat likho!")

def safe_calculate(op, x, y):
    """Safe calculator - sab operations handle karta hai"""
    try:
        if op == '+': return x + y
        elif op == '-': return x - y
        elif op == '*': return x * y
        elif op == '/':
            if y == 0:
                return "âŒ Zero se divide nahi kar sakte!"
            return x / y
        elif op == '^': return x ** y
        else:
            return "âŒ Wrong operator! (+, -, *, /, ^ use karo)"
    except:
        return "âŒ Kuch gadbad ho gaya!"

# ğŸ”¥ MAIN CALCULATOR LOOP
print("\nğŸ”¥ SHRAVAN KA SMART CALCULATOR")
print("Type 'exit' band karne ke liye")

history = []
while True:
    print("\n" + "â•"*40)
    print("1. Add (+)    2. Subtract (-)")
    print("3. Multiply (*) 4. Divide (/)")
    print("5. Power (^)    6. History")
    print("0. Exit")
    
    choice = input("Option chuno (0-6): ").strip()
    
    if choice == '0':
        print("ğŸ‘‹ Calculator band! Thanks Shravan!")
        break
    
    elif choice == '6':
        if history:
            print("\nğŸ“œ CALCULATION HISTORY:")
            for i, calc in enumerate(history[-5:], 1):
                print(f"{i}. {calc}")
        else:
            print("ğŸ“­ Koi history nahi!")
        continue
    
    elif choice in ['1','2','3','4','5']:
        print("\nâš¡ Calculation start!")
        
        # Safe numbers input
        num1 = safe_input("Pehla number: ")
        num2 = safe_input("Dusra number: ")
        
        # Operator mapping
        operators = {'1': '+', '2': '-', '3': '*', '4': '/', '5': '^'}
        op = operators[choice]
        
        # Calculate
        result = safe_calculate(op, num1, num2)
        print(f"\nâœ… {num1} {op} {num2} = {result}")
        
        # History save
        history.append(f"{num1} {op} {num2} = {result}")
        
    else:
        print("âŒ Galat option! 0-6 mein se chuno!")

print(f"\nğŸ’¾ Total {len(history)} calculations save ho gaye!")
```
