```python
print("ðŸ“Š DAY 4 - PROBLEM 4: CSV STUDENT DATABASE")
print("=" * 50)

import csv

# 1. shravan_students.csv FILE BANANA (Excel jaise)
students_data = [
    ["RollNo", "Name", "City", "Python", "Maths", "Total", "Grade"],
    [1, "Shravan", "Prayagraj", 92, 85, 177, "A+"],
    [2, "Amit", "Varanasi", 88, 78, 166, "A"],
    [3, "Rahul", "Mirzapur", 85, 92, 177, "A+"],
    [4, "Priya", "Jaunpur", 78, 88, 166, "A"]
]

# CSV FILE SAVE
with open("shravan_students.csv", "w", newline="", encoding="utf-8") as f:
    writer = csv.writer(f)
    writer.writerows(students_data)
print("âœ… shravan_students.csv SAVE ho gaya! (Excel format)")

# 2. CSV FILE PADHNA (Database jaise)
print("\nðŸ“‹ SHRAVAN KA STUDENT DATABASE:")
with open("shravan_students.csv", "r", encoding="utf-8") as f:
    reader = csv.reader(f)
    students = list(reader)
    
    # Header alag karo
    header = students[0]
    student_records = students[1:]
    
    print(f"Total Students: {len(student_records)}")
    print("RollNo | Name     | Python | Maths | Grade")
    print("-" * 45)
    
    for student in student_records:
        print(f"{student[0]:>5} | {student[1]:<8} | {student[3]:>6} | {student[4]:>5} | {student[6]}")

# 3. NEW STUDENT ADD KARNA
new_student = ["5", "Neha", "Allahabad", 90, 87, 177, "A+"]
students_data.append(new_student)

with open("shravan_students.csv", "w", newline="", encoding="utf-8") as f:
    writer = csv.writer(f)
    writer.writerows(students_data)

print("\nâž• Neha ka record ADD ho gaya!")
print("Updated count:", len(students_data))
```
