```python
print("ğŸ“š DAY 3 - PROBLEM 4: SHRAVAN LIBRARY MANAGEMENT")
print("=" * 60)

class Book:
    def __init__(self, title, author, genre, pages, price):
        self.title = title
        self.author = author
        self.genre = genre
        self.pages = pages
        self.price = price
        self.status = "Available"  # Available/Issued
    
    def display(self):
        return f"""
ğŸ“– {self.title}
   âœï¸ {self.author} | {self.genre} | {self.pages}pg
   ğŸ’° â‚¹{self.price:,.0f} | Status: {self.status}
        """.strip()
    
    def issue_book(self):
        if self.status == "Available":
            self.status = "Issued"
            return f"âœ… {self.title} ISSUED!"
        return "âŒ Book already issued!"

# ğŸ”¥ SHR AVAN'S LIBRARY - 15 HOT BOOKS
shravan_library = [
    Book("Python Crash Course", "Eric Matthes", "Programming", 544, 1200),
    Book("Clean Code", "Robert Martin", "Programming", 464, 1800),
    Book("Head First Design Patterns", "Freeman", "Programming", 691, 2200),
    Book("Automate the Boring Stuff", "Al Sweigart", "Python", 504, 1100),
    Book("Grokking Algorithms", "Aditya Bhargava", "Algorithms", 256, 900),
    Book("The Intelligent Investor", "Benjamin Graham", "Finance", 640, 850),
    Book("Atomic Habits", "James Clear", "Self-Help", 320, 450),
    Book("Sapiens", "Yuval Noah Harari", "History", 512, 650),
    Book("1984", "George Orwell", "Fiction", 328, 350),
    Book("To Kill a Mockingbird", "Harper Lee", "Fiction", 281, 400),
    Book("Rich Dad Poor Dad", "Robert Kiyosaki", "Finance", 336, 500),
    Book("Think and Grow Rich", "Napoleon Hill", "Motivation", 238, 450),
    Book("The Psychology of Money", "Morgan Housel", "Finance", 256, 600),
    Book("Deep Work", "Cal Newport", "Productivity", 304, 550),
    Book("Fluent Python", "Luciano Ramalho", "Python", 976, 2500)
]

print(f"ğŸ“š SHRAVAN LIBRARY: {len(shravan_library)} books ready!")

# ğŸ”¥ SMART LIBRARY FUNCTIONS
def search_by_genre(genre):
    return [book for book in shravan_library if genre.lower() in book.genre.lower() and book.status == "Available"]

def search_by_author(author):
    return [book for book in shravan_library if author.lower() in book.author.lower()]

def get_recommendations(budget):
    return [book for book in shravan_library if book.price <= budget and book.status == "Available"]

# ğŸ”¥ DEMO - Shravan Library Test
print("\n" + "â”€"*70)
print("ğŸ”¥ SHRAVAN LIBRARY - SMART SEARCH")
print("â”€"*70)

# 1. Programming books (Shravan ke liye perfect!)
print("\nğŸ’» Programming Books:")
programming_books = search_by_genre("programming")
for i, book in enumerate(programming_books[:4], 1):
    print(f"{i}. {book.title}")
    print(book.display())
    print()

# 2. Issue Python book
python_book = next(book for book in shravan_library if "Python" in book.title)
print(python_book.issue_book())
print(f'Python book status: {python_book.status}')

# 3. Budget recommendations (under 1000)
print("\nğŸ’° Books under â‚¹1000:")
cheap_books = get_recommendations(1000)
for book in cheap_books[:3]:
    print(f"â€¢ {book.title} - â‚¹{book.price}")

# 4. Library Summary
available_books = len([b for b in shravan_library if b.status == "Available"])
issued_books = len(shravan_library) - available_books
print(f"\nğŸ“Š SUMMARY: {available_books} Available | {issued_books} Issued")
```
