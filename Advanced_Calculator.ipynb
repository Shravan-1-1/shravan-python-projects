```python
print("üßÆ DAY 2 - PROBLEM 4: ADVANCED CALCULATOR")
print("=" * 50)

history = []

# Math functions
def add(x, y): return x + y
def subtract(x, y): return x - y
def multiply(x, y): return x * y
def divide(x, y): 
    if y == 0: return "‚ùå Can't divide by zero!"
    return x / y
def power(x, y): return x ** y
def sqrt(x):
    if x < 0: return "‚ùå Negative sqrt!"
    return x ** 0.5
def percentage(x, y): return (x / y) * 100
def factorial(n):
    if n < 0: return "‚ùå Negative factorial!"
    if n <= 1: return 1
    result = 1
    for i in range(2, int(n) + 1): result *= i
    return result

operations = {
    "1": ("Add (+)", add),
    "2": ("Subtract (-)", subtract),
    "3": ("Multiply (√ó)", multiply),
    "4": ("Divide (√∑)", divide),
    "5": ("Power (x^y)", power),
    "6": ("Square Root (‚àö)", sqrt),
    "7": ("Percentage (%)", percentage),
    "8": ("Factorial (!)", factorial)
}

while True:
    print("\n" + "="*40)
    print("üßÆ CALCULATOR MENU")
    print("1-8: Operations  |  9: History  |  0: Exit")
    
    for key, (name, _) in operations.items():
        print(f"{key}. {name}")
    
    choice = input("\nChoose (0-9): ").strip()
    
    if choice == '0':
        print("üëã Calculator closed!")
        break
        
    elif choice == '9':
        if history:
            print("\nüìú CALCULATION HISTORY:")
            for i, calc in enumerate(history[-5:], 1):  # Last 5
                print(f"{i}. {calc}")
        else:
            print("üì≠ No history yet!")
        continue
    
    elif choice in operations:
        name, func = operations[choice]
        print(f"\n‚ö° {name}")
        
        if choice == '6':  # Single arg
            try:
                x = float(input("Enter number: "))
                result = func(x)
            except:
                result = "‚ùå Invalid input!"
        elif choice == '8':  # Integer only
            try:
                x = int(input("Enter number: "))
                result = func(x)
            except:
                result = "‚ùå Enter whole number!"
        else:  # Two args
            try:
                x = float(input("First number: "))
                y = float(input("Second number: "))
                result = func(x, y)
            except:
                result = "‚ùå Invalid numbers!"
        
        print(f"‚úÖ RESULT: {result}")
        history.append(f"{name}: {x}{' ' if choice != '6' and choice != '8' else ''}{y if choice != '6' and choice != '8' else ''} = {result}")
        
    else:
        print("‚ùå Invalid choice! Try 0-9")

print(f"\nüíæ {len(history)} calculations saved in history!")
#  COMPLETE! Calculator functions + history boss level
```
