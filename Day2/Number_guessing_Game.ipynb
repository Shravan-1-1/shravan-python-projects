```python
print("ðŸŽ® DAY 2 - PROBLEM 2: NUMBER GUESSING GAME")
print("=" * 50)

import random

# Generate random number
secret = random.randint(1, 100)
attempts = 0
max_attempts = 7

print("ðŸ’¡ Guess the number between 1-100 (7 tries max)!")
print(f"Hint: It's {secret} ðŸ˜‰")  # REMOVE this line in final!

while attempts < max_attempts:
    try:
        guess = int(input(f"\nAttempt {attempts+1}/{max_attempts} - Enter guess: "))
        attempts += 1
        
        if guess == secret:
            print(f"ðŸŽ‰ WINNER! Found in {attempts} attempts! ðŸ†")
            break
        elif guess < secret:
            print("ðŸ“ˆ Too LOW! Go higher.")
        else:
            print("ðŸ“‰ Too HIGH! Go lower.")
            
    except ValueError:
        print("âŒ Enter VALID NUMBER! Try again.")
        continue

else:
    print(f"\nðŸ’¥ GAME OVER! Secret was {secret} ðŸ˜…")
    print("Practice more for Day 3! ðŸ”¥")
```
