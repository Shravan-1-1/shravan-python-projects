```python
print("üîê DAY 2 - PROBLEM 5: SECURE PASSWORD GENERATOR")
print("=" * 50)

import random
import string

def generate_password(length=12, use_upper=True, use_digits=True, use_symbols=True):
    """Generate strong password with options"""
    chars = string.ascii_lowercase
    if use_upper: chars += string.ascii_uppercase
    if use_digits: chars += string.digits  
    if use_symbols: chars += "!@#$%^&*()_+-=[]{}|;:,.<>?"
    
    if len(chars) < 3:
        return "‚ùå Need at least lowercase + 1 other type!"
    
    password = ''.join(random.choice(chars) for _ in range(length))
    
    # Ensure strength: at least 1 of each selected type
    strength_check = True
    if use_upper and not any(c.isupper() for c in password):
        strength_check = False
    if use_digits and not any(c.isdigit() for c in password):
        strength_check = False
    if use_symbols and not any(c in "!@#$%^&*()_+-=[]{}|;:,.<>?" for c in password):
        strength_check = False
        
    return password if strength_check else generate_password(length, use_upper, use_digits, use_symbols)

# Main menu
while True:
    print("\n" + "="*40)
    print("üîê PASSWORD GENERATOR")
    print("1. Quick 12-char  2. Custom  3. Bulk (5 passwords)  4. Exit")
    
    choice = input("Choose (1-4): ").strip()
    
    if choice == '1':
        pwd = generate_password(12)
        print(f"\n‚úÖ 12-CHAR PASSWORD: `{pwd}`")
        print(f"üí™ Strength: {len(pwd)} chars, all types included")
        
    elif choice == '2':
        try:
            length = int(input("Length (8-50): "))
            if 8 <= length <= 50:
                print(f"\n‚úÖ CUSTOM PASSWORD ({length} chars): `{generate_password(length)}`")
            else:
                print("‚ùå Length 8-50 only!")
        except:
            print("‚ùå Enter valid number!")
            
    elif choice == '3':
        print("\nüî• 5 STRONG PASSWORDS:")
        for i in range(5):
            pwd = generate_password(16)
            print(f"{i+1}. `{pwd}`")
            
    elif choice == '4':
        print("üîê Passwords generated successfully! Use strong ones! üëã")
        break
        
    else:
        print("‚ùå Invalid! Choose 1-4")
```
