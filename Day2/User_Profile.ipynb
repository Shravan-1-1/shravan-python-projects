```python
print("üë§ DAY 2 - PROBLEM 1: USER PROFILE CREATOR")
print("=" * 50)

# Function to get input with validation
def get_input(prompt):
    while True:
        value = input(prompt).strip()
        if value:
            return value
        print("‚ùå Empty input! Try again.")

# Main profile creator
name = get_input("Enter your full name: ")
age = int(input("Enter your age: "))
city = get_input("Enter your city (Prayagraj?): ")

# Display formatted profile
print("\n" + "="*40)
print("üìã YOUR PROFILE")
print("="*40)
print(f"Name: {name.title()}")
print(f"Age: {age}")
print(f"City: {city.title()}")
print(f"Welcome to Python Mastery, {name.split()[0]}! üéâ")

# Bonus: Age in days (rough)
days_alive = age * 365
print(f"You've lived approx {days_alive:,} days! ‚è≥")
```
