```python
print("üìö LIBRARY MANAGEMENT SYSTEM")
print("=" * 60)

# Library database
books = [
    {"id": "B001", "title": "Python Crash Course", "author": "Eric Matthes", "status": "Available", "member": ""},
    {"id": "B002", "title": "Clean Code", "author": "Robert Martin", "status": "Available", "member": ""},
    {"id": "B003", "title": "Head First Design Patterns", "author": "Freeman", "status": "Available", "member": ""},
    {"id": "B004", "title": "The Pragmatic Programmer", "author": "Hunt", "status": "Issued", "member": "Rana"},
    {"id": "B005", "title": "Grokking Algorithms", "author": "Bhargava", "status": "Available", "member": ""}
]

def display_books():
    print("\nüìñ ALL BOOKS:")
    print("-" * 80)
    print(f"{'ID':<6} {'Title':<25} {'Author':<20} {'Status':<12} {'Member':<10}")
    print("-" * 80)
    for book in books:
        print(f"{book['id']:<6} {book['title']:<25} {book['author']:<20} {book['status']:<12} {book['member']:<10}")

def search_book():
    keyword = input("\nüîç Enter book title/author: ").lower()
    print("\nüìö SEARCH RESULTS:")
    print("-" * 60)
    found = False
    for book in books:
        if keyword in book['title'].lower() or keyword in book['author'].lower():
            print(f"‚úÖ {book['id']} - {book['title']} by {book['author']} [{book['status']}]")
            found = True
    if not found:
        print("‚ùå No books found!")

def issue_book():
    book_id = input("\nüì§ Enter Book ID to issue: ").upper()
    member = input("Enter member name: ").title()
    
    for book in books:
        if book['id'] == book_id:
            if book['status'] == "Available":
                book['status'] = "Issued"
                book['member'] = member
                print(f"‚úÖ {book['title']} issued to {member}")
                return
            else:
                print(f"‚ùå {book['title']} already issued to {book['member']}")
                return
    print("‚ùå Book ID not found!")

def return_book():
    book_id = input("\nüì• Enter Book ID to return: ").upper()
    
    for book in books:
        if book['id'] == book_id:
            if book['status'] == "Issued":
                book['status'] = "Available"
                book['member'] = ""
                print(f"‚úÖ {book['title']} returned successfully!")
                return
            else:
                print(f"‚ùå {book['title']} is already available")
                return
    print("‚ùå Book ID not found!")

def stats():
    total = len(books)
    available = sum(1 for b in books if b['status'] == "Available")
    issued = total - available
    
    print("\nüìä LIBRARY STATISTICS:")
    print("-" * 40)
    print(f"Total Books:     {total}")
    print(f"Available:       {available}")
    print(f"Issued:          {issued}")
    print(f"Availability:    {(available/total)*100:.1f}%")

# Main menu
while True:
    print("\nüìã LIBRARY MENU:")
    print("1. View All Books")
    print("2. Search Books") 
    print("3. Issue Book")
    print("4. Return Book")
    print("5. Library Stats")
    print("6. Exit")
    
    choice = input("Enter choice (1-6): ")
    
    if choice == '1':
        display_books()
    elif choice == '2':
        search_book()
    elif choice == '3':
        issue_book()
    elif choice == '4':
        return_book()
    elif choice == '5':
        stats()
    elif choice == '6':
        print("\nüëã Thank you for using Library System!")
        break
    else:
        print("‚ùå Invalid choice!")
```
