```python
# üî• PROBLEM 2: SHOPPING LIST MANAGER
# PYTHON MASTERY - DAY 1

print("üõí SHOPPING CART MANAGER")
print("=" * 50)

# Shopping cart (Empty list)
cart = []
total = 0

print("\nüìù Add items (Type 'done' to finish):")
while True:
    item = input("Enter item name: ").lower()
    
    if item == 'done':
        break
        
    price = float(input(f"Enter price for {item.title()}: ‚Çπ"))
    quantity = int(input("Enter quantity: "))
    
    # Item details
    item_total = price * quantity
    
    # Add to cart
    cart.append({
        'item': item.title(),
        'price': price,
        'quantity': quantity,
        'total': item_total
    })
    
    total += item_total
    print(f"‚úÖ Added {quantity} {item.title()} - ‚Çπ{item_total}")

print("\n" + "="*50)
print("üìã YOUR SHOPPING CART:")
print("="*50)

# Display cart
if cart:
    for i, product in enumerate(cart, 1):
        print(f"{i}. {product['item']} x{product['quantity']} @ ‚Çπ{product['price']} = ‚Çπ{product['total']}")
    
    print("-" * 50)
    print(f"üí∞ GRAND TOTAL: ‚Çπ{total:.2f}")
else:
    print("üõí Cart is empty!")

print("\n‚úÖ PROBLEM 2 COMPLETE!")
print("üéØ Next: Bank Account Manager")
```
