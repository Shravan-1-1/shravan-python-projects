```python
# ðŸ”¥ PROBLEM 1 - PROFESSIONAL VERSION
print("ðŸš€ PYTHON MASTERY - DAY 1")
print("=" * 60)

num = int(input("Enter number: "))

# Calculations with superscript notation
square = num ** 2
cube = num ** 3

def factorial(n):
    if n <= 1: return 1
    fact = 1
    for i in range(2, n+1): fact *= i
    return fact

print(f"\nðŸ“Š RESULTS (Mathematical Notation):")
print(f"  {num} â†’ Square ({num}Â²) = {square}")
print(f"  {num} â†’ Cube   ({num}Â³) = {cube}")
print(f"  {num} â†’ Factorial ({num}!) = {factorial(num)}")

print(f"\nðŸ“Š RESULTS (Programming Style):")
print(f"  {num} â†’ Square ({num}^2) = {square}")
print(f"  {num} â†’ Cube   ({num}^3) = {cube}")
```
