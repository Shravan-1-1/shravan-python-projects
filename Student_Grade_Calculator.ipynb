```python
#ğŸ”¥ PROBLEM 4: Student Grade Calculator
#Features:

#Multiple subjects marks input

#Percentage calculate

#Grade assign (A+, A, B+, etc.)

#Pass/Fail status

#Grade report card
```

```python
print("ğŸ“ STUDENT GRADE CALCULATOR")
print("=" * 50)

# Subject details
subjects = ['Maths', 'Science', 'English', 'Hindi', 'Social']
total_marks = 0
total_max = 0

print("\nğŸ“š Enter marks for each subject (out of 100):")

# Input marks for all subjects
marks = {}
for subject in subjects:
    while True:
        try:
            mark = float(input(f"{subject}: "))
            if 0 <= mark <= 100:
                marks[subject] = mark
                total_marks += mark
                total_max += 100
                break
            else:
                print("âŒ Marks must be between 0-100!")
        except ValueError:
            print("âŒ Enter valid number!")

# Calculate percentage
percentage = (total_marks / total_max) * 100

# Grade calculation
def get_grade(pct):
    if pct >= 90: return "A+"
    elif pct >= 80: return "A"
    elif pct >= 70: return "B+"
    elif pct >= 60: return "B"
    elif pct >= 50: return "C"
    elif pct >= 40: return "D"
    else: return "F"

grade = get_grade(percentage)
status = "PASS" if percentage >= 40 else "FAIL"

print("\n" + "="*60)
print("ğŸ“‹ GRADE REPORT CARD")
print("="*60)

# Subject-wise display
print(f"{'Subject':<12} {'Marks':<8} {'Grade':<6}")
print("-" * 35)
for subject, mark in marks.items():
    sub_grade = get_grade((mark/100)*100)
    print(f"{subject:<12} {mark:<8.0f} {sub_grade:<6}")

print("-" * 35)
print(f"{'TOTAL':<12} {total_marks:<8.0f}/{total_max} ({percentage:.1f}%)")
print(f"{'FINAL GRADE':<12} {grade:<8} ({status})")

# Remarks
if percentage >= 90:
    remark = "Outstanding! Keep shining! ğŸŒŸ"
elif percentage >= 75:
    remark = "Excellent! Great job! ğŸ‘"
elif percentage >= 60:
    remark = "Good! Keep improving! ğŸ‘"
elif percentage >= 40:
    remark = "Pass! Work harder next time ğŸ“š"
else:
    remark = "Fail! Study harder! ğŸ’ª"

print(f"\nğŸ’¬ REMARK: {remark}")
print("\nâœ… PROBLEM 4 COMPLETE!")
print("ğŸ¯ Next: Library Book Manager")
```
