```python
# ğŸ”¥ PROBLEM 7: Temperature Converter (FINAL - DAY 1 COMPLETE)
# Features:

# Celsius â†” Fahrenheit â†” Kelvin

# Multiple conversions at once

# Unit validation

# Conversion history

# Batch converter (5 temps at once)
```

```python
print("ğŸŒ¡ï¸  TEMPERATURE CONVERTER")
print("=" * 50)

# Conversion formulas
def c_to_f(celsius):
    return (celsius * 9/5) + 32

def f_to_c(fahrenheit):
    return (fahrenheit - 32) * 5/9

def c_to_k(celsius):
    return celsius + 273.15

def k_to_c(kelvin):
    return kelvin - 273.15

def f_to_k(fahrenheit):
    return f_to_c(fahrenheit) + 273.15

def k_to_f(kelvin):
    return (kelvin - 273.15) * 9/5 + 32

history = []

def single_converter():
    try:
        temp = float(input("\nğŸŒ¡ï¸ Enter temperature: "))
        print("Convert to: 1.Celsius 2.Fahrenheit 3.Kelvin")
        to_unit = input("Choose (1-3): ")
        
        if to_unit == '1':
            print(f"ğŸ“Š {temp}Â° â†’ {temp:.1f}Â°C")
        elif to_unit == '2':
            print(f"ğŸ“Š {temp}Â° â†’ {c_to_f(temp):.1f}Â°F")
        elif to_unit == '3':
            print(f"ğŸ“Š {temp}Â° â†’ {c_to_k(temp):.1f}K")
        else:
            print("âŒ Invalid unit!")
            return
        
        history.append(f"{temp} â†’ {temp:.1f if to_unit=='1' else c_to_f(temp):.1f if to_unit=='2' else c_to_k(temp):.1f}")
        
    except ValueError:
        print("âŒ Enter valid number!")

def all_units_converter():
    try:
        celsius = float(input("\nğŸ”„ Enter temperature in Celsius: "))
        
        f = c_to_f(celsius)
        k = c_to_k(celsius)
        
        print("\nğŸ“‹ ALL UNITS:")
        print("-" * 30)
        print(f"ğŸŒ¡ï¸  Celsius:  {celsius:>6.1f}Â°C")
        print(f"ğŸ”¥ Fahrenheit: {f:>6.1f}Â°F") 
        print(f"â„ï¸  Kelvin:    {k:>6.1f}K")
        print("-" * 30)
        
        history.append(f"{celsius}Â°C â†’ {f:.1f}Â°F, {k:.1f}K")
        
    except ValueError:
        print("âŒ Enter valid number!")

def show_history():
    if history:
        print("\nğŸ“œ CONVERSION HISTORY:")
        print("-" * 40)
        for i, conv in enumerate(history[-5:], 1):  # Last 5
            print(f"{i}. {conv}")
    else:
        print("\nğŸ“­ No conversions yet!")

def batch_converter():
    temps = []
    print("\nâš¡ BATCH CONVERTER (Enter 5 temps in Celsius)")
    for i in range(5):
        try:
            temp = float(input(f"Temp {i+1}: "))
            temps.append(temp)
        except ValueError:
            print("âŒ Skipping invalid!")
    
    print("\nğŸ“Š BATCH RESULTS:")
    print(f"{'Input Â°C':<10} {'Fahrenheit':<12} {'Kelvin':<12}")
    print("-" * 35)
    for temp in temps:
        f = c_to_f(temp)
        k = c_to_k(temp)
        print(f"{temp:<10.1f} {f:<12.1f} {k:<12.1f}")

# Main menu
while True:
    print("\nğŸ”§ TEMPERATURE CONVERTER MENU:")
    print("1. Single Converter")
    print("2. All Units (Câ†’F+K)") 
    print("3. History")
    print("4. Batch (5 temps)")
    print("5. Exit")
    
    choice = input("Choose (1-5): ")
    
    if choice == '1':
        single_converter()
    elif choice == '2':
        all_units_converter()
    elif choice == '3':
        show_history()
    elif choice == '4':
        batch_converter()
    elif choice == '5':
        print("\nğŸ‰ DAY 1 COMPLETE! 7/7 PROBLEMS DONE!")
        print("ğŸ† Python Foundation Mastered!")
        print("ğŸ“… Tomorrow: DAY 2 - Lists & Loops Deep Dive")
        break
    else:
        print("âŒ Invalid choice!")
    
    input("\nâ¸ï¸ Press Enter...")
```
